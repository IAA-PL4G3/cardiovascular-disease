# Cardiovascular Disease Prognosis Demo

This notebook demonstrates how to use the modularized source code (`src/`) to load data, build features, train models, and evaluate them without duplicating code. Our source of truth is now in the Python scripts, maintaining clean architecture and reusability.

```python
import sys
import pandas as pd

# Add parent directory to path to access src module
sys.path.append('../')

from src.features.build_features import clean_data, split_and_scale
from src.models.train import train_logistic_regression, train_decision_tree, evaluate_model
```

```python
# 1. Load and Clean Data
df = pd.read_csv('../data/raw/cardio_train.csv', sep=';')
df_cleaned = clean_data(df)

print(f"Original data shape: {df.shape}")
print(f"Cleaned data shape: {df_cleaned.shape}")
```

```python
# 2. Split and Scale
X_train, X_test, y_train, y_test, scaler = split_and_scale(df_cleaned)

print(f"X_train scaled shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
# 3. Train Models
print("Training Logistic Regression...")
lr_model = train_logistic_regression(X_train, y_train)

print("Training Decision Tree...")
# Using max_depth limit to prevent severe overfitting as discovered previously
dt_model = train_decision_tree(X_train, y_train, max_depth=7)
```

```python
# 4. Evaluate Models
print("--- Logistic Regression Evaluation ---")
lr_metrics = evaluate_model(lr_model, X_test, y_test)
for metric, value in lr_metrics.items():
    print(f"{metric.capitalize()}: {value:.4f}")

print("\n--- Decision Tree Evaluation ---")
dt_metrics = evaluate_model(dt_model, X_test, y_test)
for metric, value in dt_metrics.items():
    print(f"{metric.capitalize()}: {value:.4f}")
```
